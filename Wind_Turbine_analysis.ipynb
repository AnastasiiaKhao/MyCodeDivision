```python
import pandas as pd

def get_excel_data():
  url = "https://github.com/AnastasiiaKhao/MyCodeDivision/blob/main/Wind%20Turbine%20SCADA%20Dataset%20ex.xls?raw=true"
  df = pd.read_excel(url,sheet_name="Wind Turbine SCADA Dataset")
  return df

wind_df = get_excel_data()
display(wind_df)
```

```python
wind_df['Month'] = wind_df['Date/Time'].dt.month_name()
wind_df['Year'] = wind_df['Date/Time'].dt.year
wind_df['Day'] = wind_df['Date/Time'].dt.day
wind_df['Time'] = wind_df['Date/Time'].dt.hour

def find_direction(x):
  if x>350 and x<360:
    return 'N'
  elif x > 0 and x<10:
    return 'N'
  elif x > 10 and x<30:
    return 'N/NE'
  elif x > 30 and x<50:
    return 'NE'
  elif x > 50 and x<70:
    return 'E/ENN'
  elif x > 70 and x<100:
    return 'E'
  elif x > 100 and x<120:
    return 'E/ES'
  elif x > 120 and x<140:
    return 'SE'
  elif x > 140 and x<160:
    return 'S/SE'
  elif x > 160 and x<190:
    return 'S'
  elif x > 190 and x<210:
    return 'S/SW'
  elif x > 210 and x<230:
    return 'SW'
  elif x > 230 and x<250:
    return 'W/SW'
  elif x > 250 and x<280:
    return 'W'  
  elif x > 280 and x<300:
    return 'W/NW'     
  elif x > 300 and x<320:
    return 'NW' 
  elif x > 320 and x<350:
    return 'N/NW'

wind_df['Wind Direction'] = wind_df['Wind Direction (°)'].apply(find_direction)


display(wind_df)
```

```python
wind_df["Loss_Value(kW)"]=wind_df["Theoretical_Power_Curve (KWh)"]-wind_df["LV ActivePower (kW)"]
wind_df["Loss(%)"]=wind_df["Loss_Value(kW)"]/wind_df["Theoretical_Power_Curve (KWh)"]*100

display(wind_df)
```
